```python
from hw2skeleton.cluster import cluster_by_partitioning, cluster_hierarchically, flatten, jaccard_index
from hw2skeleton.io import read_active_sites, write_clustering, write_mult_clusterings
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
active_sites = read_active_sites('data')
```

```python
partitioning_clusters = cluster_by_partitioning(active_sites,[2,3,4])
hierarchical_clusters = cluster_hierarchically(active_sites,[2,3,4])
```

```python
# plot all active sites
fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
length = list()
width = list()
height = list()
for site in active_sites:
    shape = site.get_shape()
    length.append(shape[0])
    width.append(shape[1])
    height.append(shape[2])
ax.scatter(length,width,height,marker='o')
ax.set_xlabel('Length')
ax.set_ylabel('Width')
ax.set_zlabel('Height')
plt.tight_layout()
plt.show()
```

```python
def plot_clustering(clustering):
    fig = plt.figure()
    ax=fig.add_subplot(111,projection='3d')

    color_list = ['b','r','g','orange','yellow']
    i=0
    for cluster in clustering:
        length = list()
        width = list()
        height = list()
        for site in flatten(cluster):
            shape = site.get_shape()
            length.append(shape[0])
            width.append(shape[1])
            height.append(shape[2])
        ax.scatter(length,width,height,color=color_list[i],marker='.')
        i += 1
    ax.set_xlabel('Length')
    ax.set_ylabel('Width')
    ax.set_zlabel('Height')
    plt.tight_layout()
    plt.show()
```

```python
plot_clustering(partitioning_clusters)
```

```python
plot_clustering(hierarchical_clusters)
```

```python
partitioning_clusters = cluster_by_partitioning(active_sites,[2])
hierarchical_clusters = cluster_hierarchically(active_sites,[2])
print("Jaccard Index:")
print(jaccard_index(partitioning_clusters,hierarchical_clusters))
plot_clustering(partitioning_clusters)
plot_clustering(hierarchical_clusters)
```
